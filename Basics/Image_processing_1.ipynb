```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap="gray")
                     
```

```python
show_pic(img)
```

```python
ret, img_t= cv2.threshold(img, 200,255, cv2.THRESH_BINARY)
```

```python
show_pic(img_t)
```

```python
img_t2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 6)
show_pic(img_t2)
```

```python
blended = cv2.addWeighted(img_t, 0.8, img_t2, 0.2, 0)
show_pic(blended)
```
