```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

**TASK: Create a low pass filter with a 4 by 4 Kernel filled with values of 1/10 (0.01) and then use 2-D Convolution to blur the giraffer image (displayed in normal RGB)**

```python
img = cv2.imread(r"D:/STUDY/OPENCV/drive-download-20240131T012424Z-001/Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/giraffes.jpg")
display_img(img)
```

```python
kernel = np.ones((4,4), dtype= np.float32)/10
kernel
```

```python
Convoluted_img = cv2.filter2D(img, -1, kernel)
```

```python
convoluted_img = cv2.cvtColor(Convoluted_img, cv2.COLOR_RGB2BGR)
display_img(convoluted_img)
```

**TASK: Create a Horizontal Sobel Filter (sobelx from our lecture) with a kernel size of 5 to the grayscale version of the giaraffes image and then display the resulting gradient filtered version of the image.**

```python
gray_giraffe = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
sobelx = cv2.Sobel(gray_giraffe, cv2.CV_64F, 1, 0, ksize=5)
```

```python
display_img(sobelx, cmap="gray")
```

**TASK: Plot the color histograms for the RED, BLUE, and GREEN channel of the giaraffe image. Pay careful attention to the ordering of the channels.**

```python
color= ("b", "g", "r")
i= 0
for channel in color:
    hist_values= cv2.calcHist([img], channels=[i], mask = None, histSize = [256], ranges= [0,256] )
    plt.plot(hist_values, color = channel)
    i+=1
```
